```python
import numpy as np
import random as rd
```

```python
simulations = 100000 #run 100,000 sims 
yrs = 36 # working years b4 retirement
income = 200000 #annual income
win_return = .3 # the pct's returns when your investments are positive
lose_return = -.1 # the pct's returns when your investments are negative
totals = [] #empty list to store valies
```

```python
for i in range(simulations):
    total = 0
    for j in range(yrs):
        total += income
        win = rd.randint(0,1)
        if win == 1:
            total = total * (1 + win_return)
        else: 
            total = total * (1 + lose_return)
    totals.append(total)
    
good = 0
for total in totals: 
    if total > 20125000:
        good += 1
        
print(good/simulations) # prob
print(sum(totals)/len(totals)) #overall NW after 36yrs total 
```
